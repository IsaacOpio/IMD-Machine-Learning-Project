```python
#Import libraries
from pycaret import*
import pandas as pd
import numpy as np
```

```python
data = pd.read_stata('Dataset(March2024).dta')
```

```python
data.head()
```

```python
data.describe()
```

```python
from pycaret.regression import*
```

```python
data2 = data.dropna(subset=['pc'])
```

```python
countDV = data2['pc']
print(countDV)
```

```python
s = setup(data2, target='pc', session_id=123)
```

```python
#Model training
best = s.compare_models()
```

```python
s.evaluate_model(best)
```

```python
#predict
predict_holdout =s.predict_model(best)
```

```python
new_data=data2.copy().drop('pc', axis =1)
predictions = s.predict_model(best, new_data)
```
